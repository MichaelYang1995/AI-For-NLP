```python
def func(x, k, b):
    return k * x + b
```

```python
import numpy as np
```

```python
import random
```

```python
def loss(y, y_hat):
    """
    param y: the real fares
    param y_hat: the estimated fares
    return: how good is the estimated fares

    """
    temp = zip(y, y_hat)
    # print(temp)
    result = []
    for (y1, y_hat1) in temp:
        result = result + [np.abs(y1 - y_hat1)]
    return np.mean(result)
    # return np.squre(np.abs(y - y_hat))
    # return np.sqrt(np.abs(y - y_hat))
```

```python
options_change_directions = [
   #( k,  b) 
    (+1, -1),  # k increase, b decrease
    (+1, +1), 
    (-1, -1),
    (-1, +1)   # k decrease, k increase
]
```

```python
learning_rate = 1e-1
```

```python
def step(): return  random.random() * 1
```

## 2.1 Random Choose Method to get optimal k and b

```python
min_loss_rate = float('inf')
```

```python
loop_time = 1000
```

```python
best_k = random.random() * 20 - 10
best_b = random.random() * 20 - 10
```

```python
losses = [min_loss_rate]
```

```python
while loop_time > 0:
    k_change_direction, b_change_direction = random.choice(options_change_directions)  ##每次循环都随机选择一个下降方向
    
    new_k = best_k + k_change_direction * step() * learning_rate
    new_b = best_b + b_change_direction * step() * learning_rate
    
    estimated_fares = [func(age, new_k, new_b) for age in sub_age]
    new_loss_rate = loss(sub_fares, estimated_fares)
    
    if new_loss_rate < min_loss_rate:
        min_loss_rate = new_loss_rate
        best_k = new_k
        best_b = new_b
        print('k = {}, b = {}'.format(best_k, best_b))
        print('loop = {}, min_loss_rate = {}'.format(loop_time, min_loss_rate))
        losses.append(new_loss_rate)
    loop_time -= 1
```

```python
plt.plot(range(len(losses)), losses)
plt.show()
```

## 2.2 Surprvised Direction to get optimal k and b

```python
min_loss_rate = float('inf')
```

```python
loop_time = 1000
```

```python
best_k = random.random() * 20 - 10
best_b = random.random() * 20 - 10
```

```python
direction = random.choice(options_change_directions)
```

```python
losses = [min_loss_rate]
```

```python
while loop_time > 0:
    k_change_direction, b_change_direction = direction   ##    若此下降方向可使得loss_rate变小，则继续沿此方向下降，直到无法再使loss_rate变小,
                                                         ##改变方向，直到使得loss_rate再次变小。
    new_k = best_k + k_change_direction * step() * learning_rate
    new_b = best_b + b_change_direction * step() * learning_rate
    
    estimated_fares = [func(age, new_k, new_b) for age in sub_age]
    new_loss_rate = loss(sub_fares, estimated_fares)
    
    if new_loss_rate < min_loss_rate:
        min_loss_rate = new_loss_rate
        best_k = new_k
        best_b = new_b
        print('k = {}, b = {}'.format(best_k, best_b))
        print('loop = {}, min_loss_rate = {}'.format(loop_time, min_loss_rate))
        losses.append(new_loss_rate)
    else:
        direction = random.choice(list(set(options_change_directions) - {(k_change_direction, b_change_direction)}))
    loop_time -= 1
```

```python
plt.plot(range(len(losses)), losses)
plt.show()
```

## 2.3 Gradient Descent to get optimal k and b

```python
##对loss函数求k的偏导
def derivate_k(y, y_hat, x):
    abs_values = [1 if (y_i - y_hat_i) > 0 else -1 for y_i, y_hat_i in zip(y, y_hat)]

    return np.mean([a * -x_i for a, x_i in zip(abs_values, x)])
```

```python
##对loss函数求b的偏导
def derivate_b(y, y_hat):
    abs_values = [1 if (y_i - y_hat_i) > 0 else -1 for y_i, y_hat_i in zip(y, y_hat)]
    
    return np.mean([a * -1 for a in abs_values])
```

```python
k_hat = random.random() * 20 - 10
b_hat = random.random() * 20 - 10
```

```python
loop_time = 1000
```

```python
estimated_fares = [func(age, k_hat, b_hat) for age in sub_age]
```

```python
min_loss_rate = float('inf')
```

```python
losses = [min_loss_rate]
```

```python
while loop_time > 0:
    
    k_delta = -1 * learning_rate * derivate_k(sub_fares, estimated_fares, sub_age)
    b_delta = -1 * learning_rate * derivate_b(sub_fares, estimated_fares)
    
    k_hat += k_delta
    b_hat += b_delta
    
    estimated_fares = [func(age, k_hat, b_hat) for age in sub_age]
    loss_rate = loss(sub_fares, estimated_fares)
    
    print('k = {}, b = {}'.format(k_hat, b_hat))
    print('loop = {}, min_loss_rate = {}'.format(loop_time, loss_rate))
    
    losses.append(loss_rate)
    loop_time -= 1
```

```python
plt.plot(range(len(losses)), losses)
plt.show()
```

## More Date, Better Result

```python
from collections import defaultdict
```

```python
def memo(func):
    cache = {}

    def _wrap(*args): ## ? *args, **kwargs
        if args in cache: result = cache[args]
        else:
            result = func(*args)
            cache[args] = result
        return result
    return _wrap
```

```python
@memo
def get_related_words_first(input_word, model):
    return [word for word, val in model.most_similar(input_word, topn=20)]
```

```python
def get_related_words(initial_words, model):
    unseen = initial_words
    
    seen = defaultdict(int)
    
    max_size = 500
    
    while unseen and len(seen) < max_size:
        if len(seen) % 50 == 0: 
            print('seen length : {}'.format(len(seen)))
        
        now_word = unseen.pop(0)
        
        new_append = get_related_words_first(now_word, model)
        
        unseen += new_append
        
        seen[now_word] += 1
    return seen
```

## TFDF Key words

```python
def document_frequency(word):
    return sum(1 for news_i in news_content if word in news_i)
```

```python
import math
```

```python
def idf(word):
    return math.log10(len(news_content) / document_frequency(word))
```

```python
def tf(word, one_news):
    words = one_news.split()
    
    return sum(1 for w in words if w == word)
```

```python
def get_keywords_of_a_news(one_news):
    words = set(one_news.split())
    
    tfidf=[
        (word, idf(word)*tf(word, one_news)) for word in words
    ]
    
    tfidf = sorted(tfidf, key=lambda x:x[1], reverse=True)
    return tfidf
```

```python
def search_engine(query):
    words = query.split()
    
    query_vec = vectorized.transform([' '.join(words)]).toarray()[0]
    
    words_id = [word_2_id[word] for word in words]
    
    
    news_ids = [
        set(np.where(transposed_x[id])[0]) for id in words_id
    ]
    
    result = reduce(and_, news_ids)
    
    sorted_result = sorted(result, key=lambda i: distance(query_vec, X[i].toarray()))
    
    return sorted_result
```

```python
def pretty_search_engine_with_naive_print(query):
    candidates_ids = search_engine(query)
    for i, _id in enumerate(candidates_ids):
        title = '## Search Result {}'.format(i)
        c = content['content'][_id]
        
        display(Markdown(title + '\n' + c))
```

```python
def generate_random_website():
    return 'www.' + ''.join([random.choice(ascii_uppercase) for _ in range(random.randint(3, 5))]) + '.' + random.choice(['com', 'cn', 'net'])
```

```python
generate_random_website()
```

```python
websites = [generate_random_website() for _ in range(25)]
```

```python
websites
```

```python
website_connection={
    websites[0]:random.sample(websites, 10),
    websites[1]:random.sample(websites, 5),
    websites[3]:random.sample(websites, 7),
    websites[4]:random.sample(websites, 2),
    websites[5]:random.sample(websites, 1)
}
```
